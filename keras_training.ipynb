# Preprocesing Data

```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

Example data:

```python
for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
    
for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
# print raw data
for i in train_samples:
    print(i)
```

```python
# print labels
for i in train_labels:
    print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
# print scaled data
for i in scaled_train_samples:
    print(i)
```

```python
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    Dense(16, input_shape=(1,), activation='relu'),
    Dense(32, activation='relu'),
    Dense(2, activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(Adam(lr=.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(scaled_train_samples, train_labels, validation_split=0.1, batch_size=10, epochs=20, shuffle=True, verbose=2)
```

# Preproces Test Data

```python
test_labels = []
test_samples = []
```

```python
for i in range(10):
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(0)
    
for i in range(200):
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)
    
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(1)
    
len(test_labels)
len(test_samples)
```

```python
test_labels = np.array(test_labels)
test_samples = np.array(test_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_test_samples = scaler.fit_transform((test_samples).reshape(-1,1))
```

# Predict

```python
prediction = model.predict(scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in prediction:
    print(i)
```

```python
round_prediction = model.predict_classes(scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in round_prediction:
    print(i)
```

# Confusion Matrix

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(test_labels, round_prediction)
test_labels.size
```

```python
def plot_confusion_matrix(cm, classes,
                         normalize=False,
                         title='Confusion matrix',
                         cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix, without normalization")
        
    print(cm)
    
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                horizontalalignment='center',
                color='white' if cm[i, j] > thresh else 'black')
        
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects', 'had_side_effect']
plot_confusion_matrix(cm, cm_plot_labels, title='Confusion Matrix')
```
